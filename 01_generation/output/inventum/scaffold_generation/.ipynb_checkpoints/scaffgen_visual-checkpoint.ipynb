```python
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem import AllChem

# Function to read molecules and their scores from an SDF file
def read_sdf(filename):
    suppl = Chem.SDMolSupplier(filename)
    molecules = []
    scores = []
    for mol in suppl:
        if mol is not None:
            # Compute 2D coordinates for better visualization
            AllChem.Compute2DCoords(mol) #    smiles
            molecules.append(mol)
            scores.append(float(mol.GetProp('Score')))  # Assuming 'new_NBG_score' is a property in your SDF
    return molecules, scores

# Reading molecules and their NBG scores from a file
molecules, scores = read_sdf('stage_result.sdf')

# Function to visualize molecules with NBG scores only
def visualize_molecules(molecules, scores):
    legends = [f"Score: {score:.3f}" for score in scores]
    img = Draw.MolsToGridImage(molecules, molsPerRow=5, subImgSize=(200, 200), legends=legends)
    return img

# Visualizing the molecules with NBG scores only
img = visualize_molecules(molecules, scores)
img
```

```python
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem import AllChem

# Function to read molecules and their scores from an SDF file
def read_sdf(filename):
    suppl = Chem.SDMolSupplier(filename)
    molecules = []
    scores = []
    for mol in suppl:
        if mol is not None:
            # Compute 2D coordinates for better visualization
            AllChem.Compute2DCoords(mol)
            molecules.append(mol)
            scores.append(float(mol.GetProp('Score')))  # Using 'Score' as the property
    return molecules, scores

# Reading molecules and their scores from a file
molecules, scores = read_sdf('stage_result.sdf')

# Sort molecules and scores by scores in ascending order
molecules, scores = zip(*sorted(zip(molecules, scores), key=lambda x: x[1]))

# Function to visualize molecules sorted by scores
def visualize_molecules_sorted(molecules, scores):
    legends = [f"Score: {score:.3f}" for score in scores]
    img = Draw.MolsToGridImage(molecules, molsPerRow=2, subImgSize=(400, 400), legends=legends)
    return img

# Visualizing the sorted molecules
img = visualize_molecules_sorted(molecules, scores)
img
```
